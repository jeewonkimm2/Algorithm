```python
from collections import deque

# 인접 노드 정보

graph = [
    [],
    [2,3,8],
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]
]

# 방문 유무
visited = [False] * 9
```

```python
def bfs(graph, start, visited):
  queue = deque([start])
  visited[start] = True
  while queue:
    v = queue.popleft()
    print(v, end = ' ')
    for i in graph[v]:
      if not visited[i]:
        queue.append(i)
        visited[i] = True
```

```python
bfs(graph, 1, visited)
```

```python
# jeewon's idea
jee_visited = [False] * 7

jee_graph = [
    [],
    [2,3],
    [1,4,5],
    [1,6],
    [2],
    [2],
    [3]
]
```

```python
bfs(jee_graph,1,jee_visited)
```
