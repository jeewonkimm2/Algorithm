```python
# 반복문

def factorial_iterative(n):
  result = 1
  for i in range(1, n+1):
    result *= i
  return result

print(factorial_iterative(4))
```

```python
# 재귀 : 시간 복잡도가 훨씬 좋음

def factorial_recursive(n):
  if n<=1:
    return 1
  return n * factorial_recursive(n-1)

print(factorial_recursive(4))
```
