```python
# 인접 노드 정보

graph = [
    [],
    [2,3,8],
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]
]

# 방문 유무
visited = [False] * 9
```

```python
visited
```

```python
# stack : First In Last Out

def dfs(graph, v, visited):
  visited[v] = True
  print(v, end = ' ')
  for i in graph[v]:
    if not visited[i]:
      dfs(graph, i, visited)
```

```python
dfs(graph, 1, visited)
```

```python
# Jeewon's Idea

graph = [
    [],
    [2,3],
    [1,4,5],
    [1,6],
    [2],
    [2],
    [3]
]

visited = [False] * 7
```

```python
def dfs(graph, v, visited):
  visited[v] = True
  print(v, end = ' ')
  for i in graph[v]:
    if not visited[i]:
      dfs(graph, i, visited)
```

```python
dfs(graph, 1, visited)
```
